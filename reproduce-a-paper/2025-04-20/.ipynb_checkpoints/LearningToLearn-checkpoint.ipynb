### Generate data for a one-parameter line $y=kx$

```python
# our model for data (with one parameter k)

def f1(x,k):
    
    return k*x
```

```python
# generate data 

rng = np.random.default_rng()   # a random seed to have the same data every time we run this code cell

k0 = 2  # a "true" value for k
x_upper = 4

n_pts = 10
sigma = 1   # sigma for a Gaussian noise

x_data = x_upper*rng.random(n_pts)   # x values for the "measurement"
y_true = f1(x_data,k0)   # true y values
noise = sigma*rng.normal(scale=sigma, size=n_pts)
y_data = y_true + noise   # the data = true values + the noise
xx = np.linspace(0,4,500)
yy = f1(xx, k0)

# let us record the data for future use below

data_one_par = [x_data.copy(), y_data.copy(), k0, sigma]

fig,ax = plt.subplots(ncols=1, nrows=1, figsize=(6,4))

ax.plot(xx,yy, c='darkorange', lw=2, label='truth')
ax.errorbar(x_data,y_data, yerr=sigma, fmt='b.', ms=10, label='data/measurement')

ax.grid(True,linestyle=':',linewidth='1.')
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
ax.tick_params('both',length=3,width=0.5,which='both',direction = 'in',pad=10)

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

ax.legend()
```

### Find $k$ by minimizing the deviations

We are to minimize

$$
\chi^2\equiv\sum\limits_{i=1}^N\left[y_i - k x_i\right]^2\,,
$$
with its gradient w.r.t. being

$$
\frac{\partial\chi^2}{\partial k} = -2\sum\limits_{i=1}^N x_i\left[y_i - k x_i\right] = 0 \qquad \Rightarrow \qquad k = \frac{\sum y_i x_i}{\sum x_i^2}\,.
$$

```python
k_opt = np.sum(y_data*x_data)/np.sum(x_data**2)

print('Optimal value of k: {:.5f}'.format(k_opt))
```

```python
## Sanity check using a built-in scipy function

from scipy.optimize import curve_fit

coefs,pcov = curve_fit(f1, x_data, y_data)   # pcov 

# sigmas = np.full_like(y_data, sigma)
# coefs,pcov = curve_fit(f, x_data, y_data, sigma=sigmas)
k_sp = coefs[0]

print('Scipy value of k: {:.5f}'.format(k_sp))
```

```python
## Gradient descent from scratch

num_iters = 200
lr = 0.001
k_calc = -5 + 10*rng.random()

for i in range(num_iters):
    
    k_calc += lr*np.sum(x_data*(y_data - k_calc*x_data))
    
    if i % 10 == 0:
        print(k_calc)
```

### Find $k$ with a RIM

```python
class RIMBlock(layers.Layer):
    def __init__(self, features):
        super(RIMBlock, self).__init__()
        self.features = features
        self.gru1cell = layers.GRUCell(self.features, activation='tanh')
        self.gru2cell = layers.GRUCell(self.features, activation='tanh')
        self.dense = layers.Dense(1, activation='tanh')
        
    def call(self, input_tensor, gradient, hidden_states):
        
        state1,state2 = hidden_states
        
        #X = tf.concat((input_tensor,gradient), axis=-1)

        dx,state1 = self.gru1cell(gradient,state1)
        dx,state2 = self.gru2cell(dx,state2)
        dx = self.dense(dx)
        
        return input_tensor + dx,state1,state2

class MyModel(keras.Model):
    def __init__(self, features=128, num_iters=20):
        super(MyModel, self).__init__()
        self.num_iters = num_iters
        self.features = features
        
    def build(self, input_dim):
        self.rim_step = RIMBlock(self.features)
        self.num_points = input_dim[1]
        
    def gradient(self, curr_estimate, data):
        
        X, Y = tf.unstack(data, axis=-1)
        
        return tf.reduce_sum(X*(Y - curr_estimate*X), axis=-1, keepdims=True) / self.num_points
        
    def call(self, data, return_sequence=False):
        
        state1 = tf.zeros(shape=(1,self.features))
        state2 = tf.zeros(shape=(1,self.features))

        X, Y = tf.unstack(data, axis=-1)
        output = tf.random.uniform(
            shape=[data.shape[0],1],
            minval=-5, maxval=5,
            dtype=tf.float32
        )

        loss = 0.
        if return_sequence:
            sequence = []
            for i in range(self.num_iters):
                gradient = self.gradient(output, data)
                output,state1,state2 = self.rim_step(output, gradient, [state1,state2])
                sequence.append(output)
                loss += tf.reduce_sum((Y - output*X)**2, axis=-1, keepdims=True) / self.num_points
            loss /= self.num_iters
            return tf.stack(sequence), loss
        
        for i in range(self.num_iters):
            gradient = self.gradient(output, data)
            output,state1,state2 = self.rim_step(output, gradient, [state1,state2])
            loss += tf.reduce_sum((Y - output*X)**2, axis=-1, keepdims=True) / self.num_points
        loss /= self.num_iters
        return output, loss
        
```

```python
dataset_size = 10000
batch_size = 64

x_mock_data = x_upper*rng.random(size=(dataset_size,n_pts))   # x values for the "measurement"
phi_mock = -4/8*np.pi + 4/4*np.pi*rng.random(size=(dataset_size,1))
k_mock = np.tan(phi_mock)
y_mock_true = f1(x_mock_data,k_mock)   # true y values
mock_noise = sigma*rng.normal(scale=sigma, size=(dataset_size,n_pts))
y_mock_data = y_mock_true + mock_noise   # the data = true values + the noise

data_train = tf.concat((x_mock_data[...,np.newaxis], y_mock_data[...,np.newaxis]), axis=-1)
data_train = tf.cast(data_train, dtype=tf.float32)


data_val = data_train[-2000:]
data_train = data_train[:-2000]

# # Prepare the training dataset.
train_dataset = tf.data.Dataset.from_tensor_slices(data_train)
train_dataset = train_dataset.shuffle(buffer_size=1024).batch(batch_size)

# # Prepare the validation dataset.
val_dataset = tf.data.Dataset.from_tensor_slices(data_val)
val_dataset = val_dataset.batch(batch_size)
```

```python
num_iters = 55
lr = 0.01
k_calc = -5 + 10*rng.random(size=(dataset_size,1))
k_opt = np.sum(y_mock_data*x_mock_data, axis=-1, keepdims=True)\
            / np.sum(x_mock_data**2, axis=-1, keepdims=True)

grad = np.sum(x_mock_data*(y_mock_data - k_calc*x_mock_data), axis=-1, keepdims=True)

k_sequence = [k_calc.copy()]
grad_sequence = [grad.copy()]

for i in range(num_iters):
    
    k_calc += lr*grad
    grad = np.sum(x_mock_data*(y_mock_data - k_calc*x_mock_data), axis=-1, keepdims=True)
    
    if i % 5 == 0:
        #error = np.sum((k_calc-k_mock)**2) / dataset_size
        error  = np.sum((y_mock_data - k_calc*x_mock_data)**2) / dataset_size
        #error = np.sum((k_calc-k_opt)**2) / dataset_size
        print(error)
#         fmt_string = ' '.join(['{:.4f}']*10)
#         print(fmt_string.format(*k_calc.T[0,:]))
        
    k_sequence.append(k_calc.copy())
    grad_sequence.append(grad.copy())
        
k_sequence = np.array(k_sequence)
grad_sequence = np.array(grad_sequence)
```

```python
from tqdm.notebook import tqdm

model = MyModel(features=40,num_iters=20)
optimizer = keras.optimizers.Adam(lr=0.001),
metric = keras.metrics.MeanSquaredError()

num_epochs = 10
#steps_per_epoch = 30

for epoch in range(num_epochs):

    print('Start of epoch {:d}'.format(epoch))

    train_loss = 0.
    for step,data_batch_train in tqdm(enumerate(train_dataset)):

        with tf.GradientTape() as tape:
            y_pred,curr_loss = model(
                data_batch_train,
                return_sequence=False,
                training=True
            )
            loss = tf.reduce_mean(curr_loss)

        grads = tape.gradient(loss, model.trainable_weights)
        optimizer[0].apply_gradients(zip(grads, model.trainable_weights))

        train_loss += loss
        train_loss /= step + 1

    #train_loss = metric.result()
    print('Loss over epoch: {:.8f}'.format(train_loss))
    #print("Seen so far: %s samples" % ((step + 1) * batch_size))

    val_loss = 0.
    for step,data_batch_val in enumerate(val_dataset):
        y_val,curr_loss = model(
                data_batch_val,
                return_sequence=True,
                training=False
            )
        val_loss += tf.reduce_mean(curr_loss)
    val_loss /= step + 1
    print("Validation acc: %.4f" % (float(val_loss),))
```
